## Project Objective

### What the project does
This project aims to build a **Disaster Recovery Cost Forecasting Framework** that can estimate recovery expenditures as early as possible when a disaster is declared. The framework will use historical FEMA disaster records, public assistance funding data, and relevant socio-economic and climate-related indicators to predict the likely cost of recovery for a disaster event.

The final solution is expected to go beyond prediction alone. It will also support:
- scenario-based simulations
- continuous monitoring of model performance
- automated retraining
- deployment through a FastAPI backend
- interactive visualization through a Streamlit dashboard

### Why it matters
Government agencies often struggle to estimate disaster recovery costs early enough. In many cases, recovery budgets are planned reactively, using rough estimates or historical averages rather than real-time evidence. This creates uncertainty during disaster response and recovery planning.

Because disaster recovery costs can vary significantly even for similar disaster types, early and accurate forecasting is important for better preparedness, faster response, and more reliable financial planning.

### Business value
This project creates value by helping decision-makers:
- improve early-stage budget planning for disaster recovery
- reduce the risk of underestimating or overestimating recovery costs
- support better allocation of public funds
- test “what-if” disaster scenarios under different risk conditions
- strengthen resilience planning and long-term infrastructure investment decisions

In summary, the project supports more proactive, data-driven, and financially informed disaster management.

## Problem Framing

### Type of machine learning problem
This project is primarily a **supervised machine learning regression problem** because the aim is to predict a continuous numerical outcome: disaster recovery cost.

### Possible target variable
A suitable target variable for this project is:

- **Total Obligated Recovery Cost**
- or **Total Federal Share Obligations**
- or an aggregated disaster-level cost variable derived from public assistance funded projects

The final target should represent the total cost associated with a disaster event after combining and aggregating the relevant FEMA funding records.

### Possible feature groups
The features for the model may come from multiple categories:

#### 1. Disaster declaration features
- disaster number
- state
- incident type
- declaration type
- declaration date
- incident begin date
- incident end date

#### 2. Engineered event features
- incident duration
- month or season of occurrence
- frequency of disasters in a state
- historical risk indicators
- disaster type encoding

#### 3. Financial and assistance features
- project category
- obligated amount summaries
- public assistance project counts
- historical obligation patterns

#### 4. Socio-economic enrichment features
- population
- GDP
- infrastructure index
- state-level exposure indicators

#### 5. Climate and hazard exposure features
- rainfall intensity
- wind speed
- flood severity
- other weather intensity variables if available

### Expected users of the solution
The likely users of this solution include:
- government agencies involved in disaster recovery funding
- public finance and budgeting teams
- emergency response planners
- resilience and infrastructure planning teams
- analysts and policy teams who need early cost forecasts and scenario testing tools

The model output can help these users make earlier and better-informed financial decisions during disaster events.
